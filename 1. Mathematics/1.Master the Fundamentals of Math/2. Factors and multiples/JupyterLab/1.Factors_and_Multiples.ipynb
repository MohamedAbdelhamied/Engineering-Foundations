## 1: Divisibility & Multiples

### Definitions & Rules

#### I. Divisibility Rules (The Fast-Audit Logic)
*Use these rules to determine if a number is divisible by a factor without performing long division.*
* **By 2:** Ends in an even digit (0, 2, 4, 6, 8).
* **By 3:** The sum of the digits is divisible by 3.
* **By 4:** The last two digits are divisible by 4.
* **By 5:** Ends in 0 or 5.
* **By 6:** Divisible by **both** 2 and 3.
* **By 7 (The Extraction Rule):** Take the last digit, multiply it by 5, and add it to the remaining digits. If the sum is divisible by 7, the original number is too.
* **By 8:** The last three digits are divisible by 8.
* **By 9:** The sum of the digits is divisible by 9.
* **By 10:** Ends in 0.

#### II. Multiples (The Scaling Logic)
*A multiple is the product of a given number and any integer.*
* **Formula:** $n \times \{1, 2, 3, 4, \dots\}$
* **Key Property:** Multiples form an infinite set where elements are always $\ge$ the original number.
* **Critical Check:** If $x$ is a multiple of $y$, then $y$ must be a factor of $x$.

### Python Implementation

```python
def check_divisibility(number):
    """
    Validates a number against fundamental divisibility rules (2 through 10)
    and demonstrates step-by-step logic checks.
    """
    results = {}
    str_num = str(number)
    digits = [int(d) for d in str_num]
    
    # Rule 2
    results[2] = (number % 2 == 0)
    
    # Rule 3
    digit_sum = sum(digits)
    results[3] = (digit_sum % 3 == 0)
    
    # Rule 4
    last_two = int(str_num[-2:]) if len(str_num) >= 2 else number
    results[4] = (last_two % 4 == 0)
    
    # Rule 5
    results[5] = (str_num[-1] in ['0', '5'])
    
    # Rule 6
    results[6] = (results[2] and results[3])
    
    # Rule 7 (Extraction Rule)
    if len(str_num) > 1:
        rest, last = int(str_num[:-1]), digits[-1]
        extracted_sum = rest + (last * 5)
        results[7] = (extracted_sum % 7 == 0)
    else:
        results[7] = (number % 7 == 0)
        
    # Rule 8
    last_three = int(str_num[-3:]) if len(str_num) >= 3 else number
    results[8] = (last_three % 8 == 0)
    
    # Rule 9
    results[9] = (digit_sum % 9 == 0)
    
    # Rule 10
    results[10] = (str_num[-1] == '0')
    
    print(f"--- Divisibility Audit for: {number} ---")
    print(f"Digit Sum (for 3 & 9 Verification): {digit_sum}")
    for rule, passed in results.items():
        status = "PASSED" if passed else "FAILED"
        print(f"Divisibility by {rule}: {status}")
    return results

def generate_multiples(number, count=5):
    """
    Generates a set of sequential scaling multiples for a base number.
    """
    multiples = [number * i for i in range(1, count + 1)]
    print(f"\n--- Multiples of {number} (First {count}) ---")
    print(f"{number} * {{1, 2, 3, 4, 5...}} -> {multiples}")
    return multiples

# Execution Check
div_audit = check_divisibility(153)
multiples_audit = generate_multiples(7, 5)
```

## 2: Prime & Composite Numbers

### Definitions & Rules

#### I. Prime Numbers (The "Unbreakables")
* A whole number greater than 1 that has exactly two factors: **1 and itself**.
* **Key Sequence:** 2, 3, 5, 7, 11, 13, 17, 19, 23...
* **The Single Exception:** **2** is the only even prime number. All other even numbers are divisible by 2, classifying them as composite.
* **The Zero/One Void:** Neither 0 nor 1 is prime. 1 fails the definition because it only has one distinct factor (itself).

#### II. Composite Numbers (The "Constructed")
* A whole number greater than 1 that has **more than two factors**.
* These numbers can be broken down completely into smaller prime building blocks.

#### III. Identification Strategy
1. **Check Evenness:** If the value ends in an even digit and is not 2, it is immediately **Composite**.
2. **Apply Shorthand Tests:** Evaluate digit sum for 3/9, or trailing digits for 5.
3. **The Target Gate:** The moment a single factor other than 1 and itself is verified, terminate exploration: it is Composite.

### Python Implementation

```python
def classify_number(n):
    """
    Classifies a number as Prime or Composite using defensive optimization rules.
    """
    if n <= 1:
        return f"{n} is neither Prime nor Composite."
    if n == 2:
        return "2 is PRIME ( Factors are: [1, 2] )--The Unique Even Prime-- "
    if n % 2 == 0:
        return f"{n} is COMPOSITE (Even number check: divisible by 2, factors include: 1, 2, {n})"
        
    # Standard Primality Loop up to square root for optimization
    factors = [1]
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            factors.append(i)
            if i != n // i:
                factors.append(n // i)
            factors.append(n)
            factors.sort()
            # Clean deduplicated list of discovered factors
            discovered = sorted(list(set(factors)))
            return f"{n} is COMPOSITE (Factors are: [{', '.join(map(str, discovered))}])"
            
    return f"{n} is PRIME (Factors are: [1, {n}])"

# Execution Check
print("--- Primality Verification ---")
print(classify_number(1))
print(classify_number(2))
print(classify_number(9))
print(classify_number(23))
```

## 3: Prime Factorization

### Definitions & Rules

#### I. Core Methodology: The Factor Tree
* **The Fundamental Theorem:** Every composite number can be uniquely expressed as a product of prime powers.
* **The Splitting Trace:** Split the composite target into any two recognizable factors. If a branch node hits a prime, circle/box it—that line of execution terminates. If it is composite, split recursively until all endpoints are prime.

#### II. Writing Output (Product of Primes)
* **Standard Representation:** Arrange prime nodes chronologically from smallest to largest with multiplication symbols ($\cdot$).
* **Exponential Compaction:** Condense matching bases into exponential format ($x^n$) to streamline future matrix calculations.

### Python Implementation


```python
def get_prime_factorization(n):
    """
    Computes prime factorization and prints an explicit trace log
    reconstructing the factor tree branches step-by-step.
    """
    if n <= 1:
        return {}
    
    target = n
    factors = []
    divisor = 2
    steps = []
    
    while target > 1:
        # Check if the current divisor splits the remaining composite dividend
        while target % divisor == 0:
            factors.append(divisor)
            next_balance = target // divisor
            
            steps.append(f"  [Prime Factor Found]: {divisor} | [Composite Dividend Balance]: {next_balance}")
            target = next_balance
            
        divisor += 1
        # Optimization check: if the remaining target is smaller than divisor squared,
        # the leftover target is prime, terminating the tree.
        if divisor * divisor > target and target > 1:
            factors.append(target)
            steps.append(f"  [Terminal Leaf Node]: {target} (No further splits possible; Prime terminal)")
            break

    # Construct Exponential Forms
    unique_factors = sorted(list(set(factors)))
    exponential_notation = " * ".join([f"{f}^{factors.count(f)}" if factors.count(f) > 1 else f"{f}" for f in unique_factors])
    product_of_primes = " * ".join([str(f) for f in factors])
    
    print(f"--- Factor Tree Execution for: {n} ---")
    for step in steps:
        print(step)
    print(f"\nProduct of Primes: {product_of_primes}")
    print(f"Exponential Form:  {exponential_notation}\n")
    
    return {f: factors.count(f) for f in unique_factors}

# Execution Check (Verification Examples)
f1 = get_prime_factorization(102)
f2 = get_prime_factorization(84)
f3 = get_prime_factorization(80)
f4 = get_prime_factorization(147)
```

## 4: Least Common Multiple (LCM)

### Definitions & Rules

#### I. Method 1: Brute Force Listing (Small Sets)
* Write sequential multiples for each term until the absolute lowest shared element intersects both vectors.

#### II. Method 2: Prime Factorization Analysis (The Global Maximum Rule)
* When sets are large, brute force fails. Factor all target numbers into prime exponential formats. Analyze the prime bases across all targets and extract the **highest power** present for every unique base. Multiply these maximums together.

#### III. Monomial Algebraic Expansion (Variables)
* Isolate and solve the numeric factors first. For variable string streams ($a^n b^m$), treat each letter base like a prime entry: select the **highest exponent** found across the inputs. Combine the results.

#### ⚠️ The Architecture Warning
* Never calculate an LCM by blindly multiplying target items together ($x \cdot y \cdot z$). Attempting to shortcut $4, 5, 10$ via brute multiplication outputs $200$, violating the true mathematically optimized minimum of $20$.

### Python Implementation

```python
def calculate_lcm_numeric(numbers, log_prefix="  "):
    """
    Internal helper computing numeric LCM using the Global Maximum Rule with explicit logging.
    """
    from collections import defaultdict
    global_prime_max = defaultdict(int)
    
    print(f"{log_prefix}--- Step 1: Numeric Coefficient Analysis for {numbers} ---")
    for num in numbers:
        target = num
        current_factors = defaultdict(int)
        divisor = 2
        while target > 1:
            while target % divisor == 0:
                current_factors[divisor] += 1
                target //= divisor
            divisor += 1
            if divisor * divisor > target and target > 1:
                current_factors[target] += 1
                break
        
        print(f"{log_prefix}Coeff {num} Prime Decomposition: " + 
              ", ".join([f"{k}^{v}" for k, v in current_factors.items()]) if current_factors else f"{log_prefix}Coeff {num} Prime Decomposition: 1")
        
        for prime, power in current_factors.items():
            if power > global_prime_max[prime]:
                global_prime_max[prime] = power
                
    lcm_result = 1
    calculation_string = []
    for prime, power in global_prime_max.items():
        lcm_result *= (prime ** power)
        calculation_string.append(f"{prime}^{power}")
        
    print(f"{log_prefix}Selected Numeric Maximums: " + (", ".join(calculation_string) if calculation_string else "1"))
    print(f"{log_prefix}Evaluated Numeric LCM: {lcm_result}")
    return lcm_result

def calculate_monomial_lcm(mono1, mono2):
    """
    Computes LCM for algebraic monomials, displaying explicit variable comparison matrices.
    Monomial format: (coefficient, {'variable': exponent})
    """
    num1, vars1 = mono1
    num2, vars2 = mono2
    
    # Format representation strings for logging
    str_m1_vars = "".join([f"{k}^{v}" if v > 1 else f"{k}" for k, v in sorted(vars1.items())])
    str_m2_vars = "".join([f"{k}^{v}" if v > 1 else f"{k}" for k, v in sorted(vars2.items())])
    print(f"--- Algorithmic Monomial LCM Analysis for: {num1}{str_m1_vars} and {num2}{str_m2_vars} ---")
    
    # 1. Numeric LCM Section
    lcm_numeric = calculate_lcm_numeric([num1, num2], log_prefix="  ")
    
    # 2. Variable Maximum Matrix Section
    print(f"  --- Step 2: Variable Power Comparison Matrix (Global Maximum Rule) ---")
    all_vars = set(vars1.keys()).union(set(vars2.keys()))
    result_vars = {}
    
    for v in sorted(all_vars):
        p1 = vars1.get(v, 0)
        p2 = vars2.get(v, 0)
        max_p = max(p1, p2)
        result_vars[v] = max_p
        print(f"    Variable [{v}] Exponents -> Term 1: {p1}, Term 2: {p2} | Selected Max Power: {v}^{max_p}")
        
    # 3. Final Assembly
    var_string = "".join([f"{k}^{v}" if v > 1 else f"{k}" for k, v in sorted(result_vars.items()) if v > 0])
    final_output = f"{lcm_numeric}{var_string}"
    print(f"\n  Final Assembled Monomial LCM Result: {final_output}\n")    

# Execution Check - Standard Numbers Only
print("--- Standard Numeric LCM Verification ---\n")
calculate_lcm_numeric([24, 10, 15], log_prefix="")
print("\n"+ "*"*70+"\n")
print("--- System Flaw Defensive Check ---\n")
calculate_lcm_numeric([4, 5, 10], log_prefix="") # Expected: 20
print("\n"+ "*"*70+"\n")
# Execution Check - Monomial Analysis (15a^2b and 10ab^3)
m1 = (15, {'a': 2, 'b': 1})
m2 = (10, {'a': 1, 'b': 3})
calculate_monomial_lcm(m1, m2)
```

## 5: Greatest Common Factor (GCF)

### Definitions & Rules

#### I. Core Methodology: The Common Minimum Rule (Intersection Matrix)
* Unlike LCM which collects maximum instances across all bases, the Greatest Common Factor (GCF) isolates only the prime bases **shared comprehensively by all terms**, and extracts the **lowest power** common to them.

#### II. Algebraic Processing Steps
1. Break all base coefficients down into core prime products.
2. Fully expand literal string variables down to individual elements ($x^3 \rightarrow x \cdot x \cdot x$).
3. Identify and map the intersection set containing elements found across **every single entry**.
4. Multiply this intersection set together to isolate the true GCF.

### Python Implementation

```python
def calculate_gcf_numeric(numbers, log_prefix="  "):
    """
    Internal helper computing numeric GCF using the Common Minimum Intersection Rule with explicit logging.
    """
    from collections import defaultdict
    prime_clones = []
    all_primes = set()
    
    print(f"{log_prefix}--- Step 1: Numeric Coefficient Intersection for {numbers} ---")
    for num in numbers:
        target = num
        current_factors = defaultdict(int)
        divisor = 2
        while target > 1:
            while target % divisor == 0:
                current_factors[divisor] += 1
                all_primes.add(divisor)
                target //= divisor
            divisor += 1
            if divisor * divisor > target and target > 1:
                current_factors[target] += 1
                all_primes.add(target)
                break
        prime_clones.append(current_factors)
        print(f"{log_prefix}Term {num} Primes Decomposition: " + 
              ", ".join([f"{k}^{v}" for k, v in current_factors.items()]) if current_factors else f"{log_prefix}Term {num} Primes Decomposition: 1")

    global_gcf_map = {}
    for prime in sorted(all_primes):
        powers = [entry.get(prime, 0) for entry in prime_clones]
        min_power = min(powers)
        if min_power > 0:
            global_gcf_map[prime] = min_power

    gcf_result = 1
    calculation_string = []
    for prime, power in global_gcf_map.items():
        gcf_result *= (prime ** power)
        calculation_string.append(f"{prime}^{power}")
        
    if not global_gcf_map:
        calculation_string = ["1 (No shared prime elements)"]
        gcf_result = 1

    print(f"{log_prefix}Selected Common Numeric Minimums: " + ", ".join(calculation_string))
    print(f"{log_prefix}Evaluated Numeric GCF: {gcf_result}")
    return gcf_result

def calculate_monomial_gcf(mono_list):
    """
    Computes GCF across an arbitrary list of complex algebraic monomials, logging the variable tracking matrix.
    Monomial representation: (coefficient, {'variable': exponent})
    """
    # Format representation strings for group display logging
    display_terms = []
    for num, vars_dict in mono_list:
        v_str = "".join([f"{k}^{v}" if v > 1 else f"{k}" for k, v in sorted(vars_dict.items())])
        display_terms.append(f"{num}{v_str}")
    
    print(f"--- Algorithmic Monomial GCF Analysis for Group: {', '.join(display_terms)} ---")
    
    # 1. Numeric GCF Section
    coefficients = [item[0] for item in mono_list]
    numeric_gcf = calculate_gcf_numeric(coefficients, log_prefix="  ")
    
    # 2. Variable Minimum Intersection Matrix Section
    print(f"  --- Step 2: Variable Power Intersection Matrix (Common Minimum Rule) ---")
    
    # Collect all unique variables used across all terms
    all_seen_vars = set()
    for item in mono_list:
        all_seen_vars.update(item[1].keys())
        
    final_vars = {}
    for var in sorted(all_seen_vars):
        # Extract exponents for this variable across every item in the list
        powers = [item[1].get(var, 0) for item in mono_list]
        min_power = min(powers)
        
        # Build an explicit visual log of the matrix row
        trace_powers_str = ", ".join([f"Term {i+1}: {p}" for i, p in enumerate(powers)])
        print(f"    Variable [{var}] Exponents -> {trace_powers_str} | Selected Min Intersection: {var}^{min_power}")
        
        if min_power > 0:
            final_vars[var] = min_power
        
    # 3. Final Assembly
    var_string = "".join([f"{k}^{v}" if v > 1 else f"{k}" for k, v in sorted(final_vars.items()) if v > 0])
    final_output = f"{numeric_gcf}{var_string}"
    print(f"\n  Final Assembled Monomial GCF Result: {final_output}\n")    

# Execution Check - Pure Numerical
print("--- Standard Numeric GCF Verification ---\n")
calculate_gcf_numeric([5, 10, 20], log_prefix="")
print("\n"+ "*"*70+"\n")
# Execution Check - Single Variable Monomials (6a, 6a^2, 12a^3)
mono_set1 = [
    (6, {'a': 1}),
    (6, {'a': 2}),
    (12, {'a': 3})
]
calculate_monomial_gcf(mono_set1)
print("\n"+ "*"*70+"\n")
# Execution Check - Multi-Variable Monomials (4x^2y, 2xy, 10xy^2)
mono_set2 = [
    (4, {'x': 2, 'y': 1}),
    (2, {'x': 1, 'y': 1}),
    (10, {'x': 1, 'y': 2})
]
calculate_monomial_gcf(mono_set2)
```
