# Lesson 2 & 3: The Symmetry (Identity & Opposites)

## 1. Identity Property
**Definition:** An Identity Number is a value that, when applied in an operation, leaves the other number unchanged.

* **Additive Identity ($0$):** $x + 0 = x$
* **Multiplicative Identity ($1$):** $x \cdot 1 = x$

---

## 2. Opposite Numbers (Additive Inverse)
**Definition:** The "Opposite" is the mirror image of a number across the zero-axis on a number line. Adding a number to its opposite always returns the **Additive Identity (0)**.

**Mathematical Rule:** $a + (-a) = 0$

```python
import matplotlib.pyplot as plt

# Use default clean white background
plt.style.use('default')

# 1. COMPACT SIZE: Reducing width and height to prevent 'stretching'
fig, ax = plt.subplots(figsize=(7, 1.2)) 

# 2. Main Line with Arrows
ax.annotate('', xy=(3.8, 0), xytext=(-3.8, 0),
            arrowprops=dict(arrowstyle='<->', color='black', lw=1))

# 3. Plotting markers and Magnetic Labels
all_points = range(-3, 4)
for x in all_points:
    # Vertical tick mark
    ax.plot([x, x], [-0.03, 0.03], color='black', lw=1)
    
    # Label logic
    if x == -3:
        label, color = "-3\n(Opposite)", 'red'
    elif x == 0:
        label, color = "0\n(Identity)", 'blue'
    elif x == 3:
        label, color = "3\n(Original)", 'green'
    else:
        label, color = str(x), 'black'
    
    # Attach labels directly to ticks
    ax.text(x, -0.06, label, ha='center', va='top', fontsize=9, fontweight='bold', color=color)

# 4. Key colored dots
ax.scatter([-3, 0, 3], [0, 0, 0], color=['red', 'blue', 'green'], s=50, zorder=3)

# 5. LOWER PURPLE LINE: 'rad=.2' keeps the arc very low to the line
ax.annotate('', xy=(-3, 0.04), xytext=(3, 0.04),
            arrowprops=dict(arrowstyle="<->", connectionstyle="arc3,rad=.1", 
            color='rebeccapurple', lw=2))

# 6. TIGHT CONSTRAINTS: Narrower x-limits make the line look more compact
ax.set_xlim(-4, 4)
ax.set_ylim(-0.7, 0.4)
ax.axis('off')

# Title with smaller padding
plt.title("Visualizing Mathematical Symmetry & Identity", fontsize=11, fontweight='bold', pad=4)

plt.show()
```

**Bridge 1 & 2: Software Architecture & Linux (Defaults)**
* Logic: $0$ and $1$ are the "Factory Resets" of a system.

```python
# Software Architecture: Default State Logic
def process_data(value, increment=0, multiplier=1):
    # If no values are passed, the Identity properties ensure the data is unchanged
    return (value + increment) * multiplier

original_data = 100
print(f"Original: {original_data} | Result with Identity Defaults: {process_data(original_data)}")
```

**Bridge 3: Computer Architecture (Zeroing Registers)**
* Logic: Adding an opposite to clear memory.

```python
# CPU Register Logic: Zeroing out
register_a = 255  # Active data
opposite_val = -register_a
clear_operation = register_a + opposite_val

print(f"Register State: {clear_operation} (System Ready for new task)")
```

**Bridge 4: Cybersecurity (Anomaly Offsets)**
* Logic: Using Identity to find "zero-day" changes.

```python
# Security: Baseline Verification
baseline_signature = 1024
current_signature = 1024

if (current_signature + (-baseline_signature)) == 0:
    print("Security Status: INTEGRITY VERIFIED (Identity State)")
else:
    print("Security Status: ANOMALY DETECTED")
```

**Bridge 5 & 6: AI & Cloud Computing (Scaling)**
* Logic: Multiplicative Identity ($1$) as the "Base Scale" for Cloud instances.

```python
# Cloud Scaling: Multiplicative Identity
base_instance_cost = 50.0
scale_factor = 1 # The Identity scale (No change)
final_cost = base_instance_cost * scale_factor

print(f"Cloud Cost: ${final_cost} (Base Identity)")
```
