# Lesson 4 & 5: Operations with Signed Numbers (Multiplication & Division)

## 1. The Unified Matrix
The core principle for determining the resulting sign is identical for both **Multiplication (•)** and **Division (÷)**. The logic depends entirely on whether the signs are identical or opposing.

| Operation | Sign Relationship | Resulting Sign |
| :--- | :--- | :--- |
| $(+) \cdot (+)$ or $(+) \div (+)$ | **Same Signs** | **$(+)$ Positive** |
| $(-) \cdot (-)$ or $(-) \div (-)$ | **Same Signs** | **$(+)$ Positive** |
| $(+) \cdot (-)$ or $(+) \div (-)$ | **Different Signs** | **$(-)$ Negative** |
| $(-) \cdot (+)$ or $(-) \div (+)$ | **Different Signs** | **$(-)$ Negative** |

---

## 2. General Rules
* **The Identity Rule:** Any number multiplied or divided by its same sign results in a **Positive**.
* **The Conflict Rule:** Any number multiplied or divided by an opposing sign results in a **Negative**.

## 3. Multiple Factor Rule (Chain Operations)
When processing a sequence of multiple terms (e.g., $(+2)(-3)(+4)(-5)$):
1.  **Count the total number of Negative $(-)$ signs.**
2.  **Even Count:** The final result is **Positive $(+)$**.
3.  **Odd Count:** The final result is **Negative $(-)$**.

## 🛡️ The 6-Bridge Engineering Framework

### Bridge 1: Linux/Kernel (Bitwise Permissions)
In Linux, file permissions (Read=4, Write=2, Execute=1) are calculated using logic similar to signed multiplication. When we combine bits, we are essentially performing a "sign-check" on the permission flags.

```python
# Simulating a permission check (Bitwise AND acts as a sign gate)
REQUIRED_PERMISSION = 0b100 # Read (4)
USER_PERMISSION = 0b110     # Read + Write (6)

if (USER_PERMISSION & REQUIRED_PERMISSION):
    print("Access Granted: Permission Flag is Positive.")
else:
    print("Access Denied: Permission Flag is Zero/Negative.")
```

### Bridge 2: Cybersecurity (XOR Stream Ciphers)
The "Unified Matrix" is the human version of an **XOR Gate**. In cryptography, XORing a plaintext bit with a key bit follows the same "Same = Positive/0, Different = Negative/1" logic, which is the foundation of the One-Time Pad.

```python
# XOR Logic: Identical = 0, Different = 1
def xor_cipher_logic(bit_a, bit_b):
    return bit_a ^ bit_b

print(f"Same Bits (1,1): {xor_cipher_logic(1,1)}") # Result 0 (Like Positive)
print(f"Different Bits (1,0): {xor_cipher_logic(1,0)}") # Result 1 (Like Negative)
```

### Bridge 3: CPU Architecture (The Sign Flag SF)
CPUs use a specific register called the **Sign Flag**. When an operation results in a negative number, the SF is set to 1. This is calculated by checking the Most Significant Bit (MSB).

```python
def check_sign_flag(value):
    # Simulating an 8-bit signed integer check
    # If the 8th bit is 1, it's negative
    sf = 1 if value < 0 else 0
    return f"Value: {value} | Sign Flag (SF): {sf}"

print(check_sign_flag(-15))
print(check_sign_flag(25))
```

### Bridge 4: Software Architecture (Input Validation Decorators)
Architects use sign-logic to enforce business rules. For example, a financial system must ensure that "Quantity" is never multiplied by a negative "Price" unless a "Refund" flag is present.

```python
def validate_transaction(price, quantity):
    # Rule: Price * Quantity must follow logical sign constraints
    total = price * quantity
    status = "REFUND" if total < 0 else "PURCHASE"
    return f"Total: {total} | Transaction Type: {status}"

print(validate_transaction(-50.0, 2))
```

### Bridge 5: AI & Machine Learning (Weight Activation)
In Neural Networks, a "Neuron" multiplies an input by a weight. If both are negative, the activation becomes positive ($(-1) \cdot (-1) = 1$). This allows the network to learn that "two negatives make a positive" feature.

```python
input_signal = -0.85
synapse_weight = -1.2

activation = input_signal * synapse_weight
print(f"Neuron Activation: {activation:.2f} (Positive reinforcement from two negatives)")
```

### Bridge 6: Cloud Infrastructure (Cost Offsets)
Cloud billing often uses negative multipliers for "Reserved Instance Credits." Multiplying a negative unit price by hours used creates a "Negative Cost" (Credit) that offsets the total bill.

```python
def cloud_billing(units, unit_price):
    bill = units * unit_price
    return f"Total Cloud Charge: ${bill}"

# Multiplying hours by a negative credit rate
print(cloud_billing(720, -0.05))
```
