## 1. System Access & Privilege Elevation
## ---------------------------------------------------------------------------

### 🛡️ The Principle of Least Privilege
Standard workflow involves operating as a regular user (`kali` or `parrot`) and elevating only when required. This prevents accidental system-wide changes or unintended execution of malicious scripts with full authority.

### 🔑 Key Concepts & Commands
* **sudo (SuperUser Do):** Temporary elevation for a single command.
* **su (Switch User):** Transitioning sessions. 
    * `sudo su -`: The `-` ensures a full login shell, loading the target user's (root) environment variables.
* **Isolation:** Privilege elevation is process-specific. A `sudo` session in one terminal tab does not grant privileges to other open tabs.

### 📂 The "Gatekeeper" Test
Testing access to the password hash file to verify privilege state:
* `cat /etc/shadow` -> **Result:** Permission Denied (System Integrity check).
* `sudo cat /etc/shadow` -> **Result:** Access Granted (Requires password).

> **Empire Note:** In 2020, Kali shifted from default root to the `sudo` model to align with modern security standards.

## 2. Linux Filesystem Navigation
## ---------------------------------------------------------------------------

### 📍 Understanding the Prompt
The terminal prompt `kali@kali:~$` breaks down as:
* **User:** `kali` (The current logged-in identity)
* **Hostname:** `@kali` (The name of the machine)
* **Location:** `~` (Tilde represents the Home directory: `/home/kali`)

### 🧭 Navigation Essentials
* `pwd`: Print Working Directory. Shows exactly where you are.
* `cd [path]`: Change Directory.
    * `cd ..`: Move up one level.
    * `cd ~`: Jump straight back to the Home directory.
    * `cd /`: Move to the system Root.
* **Tab Completion:** Use `Tab` to autocomplete folder names; hit `Tab` twice to show all possibilities.

### 🔍 Listing & Manuals
* `ls`: List files and folders.
* `ls -la`: The "Pro" view.
    * `-l`: Long format (shows permissions, owner, size).
    * `-a`: Show all (includes hidden files starting with a `.`).
* `man [command]`: Opens the manual page (Use `q` to quit).
* `[command] --help`: Quick reference for command flags.
* **Resource:** [explainshell.com](https://explainshell.com) - Visual breakdown of complex commands.

### 📁 File & Directory Operations
* `mkdir [name]`: Create a directory.
* `rmdir [name]`: Remove an **empty** directory.
* `rm [file]`: Remove a file.
* `cp [source] [destination]`: Copy a file.
* `mv [source] [destination]`: Move or Rename a file.

### ⚡ Power User Tools
* `ctrl + l`: Clear the terminal screen (same as typing `clear`).
* `echo 'text' > file.txt`: Redirects output to create/overwrite a file.
* `cat file.txt`: Concatenate and print file content to the screen.
* `sudo updatedb`: Updates the system file database (required for `locate`).
* `locate [filename]`: Quickly find the path of a file.
* `passwd`: Change the current user's password.

> **Crucial Note:** Linux is **Case Sensitive**. `Downloads` and `downloads` are two different things.

## 3. User & Privilege
## ---------------------------------------------------------------------------
### 1. Permissions Management
Use these to fix "Permission Denied" errors or secure sensitive files you discover.

```bash
# Give a script execution rights
chmod +x script_name.sh

# Secure sensitive loot (Read/Write only for the owner)
chmod 600 sensitive_file.txt

# Set standard directory permissions (Owner: all, Others: read/execute)
chmod 755 directory_name/
```

### 2. User and Group Discovery
Use these to identify the environment's security posture and user list.

```bash
# List all system users and their shells
cat /etc/passwd

# Check what elevated commands your current user can run
sudo -l

# Check which users belong to the 'sudo' group
grep 'sudo' /etc/group
```

### 3. Switching Contexts
Use these when you need to pivot between a low-privilege user and an administrative role.

```bash
# Switch to a specific user (e.g., john)
su john

# Elevate to the root user session
sudo su -
```

## 4.🌐 Service & Server Management
## ---------------------------------------------------------------------------
### 1. System Service Control
Use these to manage background tools like the SSH server or Apache web server.

```bash
# Check if a service (like apache2) is currently running
systemctl status apache2

# Start a service manually
sudo systemctl start apache2

# Stop a service manually
sudo systemctl stop apache2

# Set a service to start automatically every time you boot Linux
sudo systemctl enable apache2
```

### 2. Instant Web Hosting (Python)
Use this to serve files from your current directory over the network instantly.

```bash
# Start a web server on port 80 (standard HTTP)
# Note: Ctrl+C in the terminal will stop this server
sudo python3 -m http.server 80
```

### 3. Web Directory Navigation
Quick access to the default directory for hosted web content.

```bash
# Move to the web root directory
cd /var/www/html/

# List contents to see hosted files
ls -la
```

## 5.📜 Bash Scripting Template & Automation
## ---------------------------------------------------------------------------
### 1. The Basic Script Structure
Use this as the starting point for every new tool you build.

```bash
#!/bin/bash

# Define variables
TARGET="google.com"

# Output information to the user
echo "Starting recon on $TARGET"
```

### 2. Handling User Input (Arguments)
This allows you to make your scripts dynamic so you don't have to edit the code every time you change a target.

```bash
# $1 represents the first word you type after the script name
# Usage: ./myscript.sh 192.168.1.1
TARGET_IP=$1

echo "Targeting IP: $TARGET_IP"
```

### 3. Logic and Error Handling
This `if` statement ensures the script doesn't run if you forget to provide a target.

```bash
if [ "$1" == "" ]
then
    echo "Error: You forgot the target!"
    echo "Usage: ./script.sh [IP_ADDRESS]"
else
    echo "Target $1 accepted. Proceeding..."
fi
```

### 4. The Automation Loop (Scanner Engine)
This is the logic used in your `ip_sweep.sh`. It runs a command for every number in a sequence.

```bash
# Scans 192.168.1.1 through 192.168.1.254
for ip in $(seq 1 254); do
    ping -c 1 192.168.1.$ip | grep "64 bytes" &
done
```
