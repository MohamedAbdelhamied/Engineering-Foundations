## 1. The Rule of Vectors
In system logic, we treat operations as directional instructions:
* **Addition (+):** Continue in the direction of the sign.
* **Subtraction (-):** Flip the current direction.

**Double Negative Logic:** Subtracting a negative is a "double flip," effectively becoming addition.
* **Mathematical Rule:** $a - (-b) = a + b$
* **Example:** $5 - (-3) \Rightarrow 5 + 3 = 8$

```python
import matplotlib.pyplot as plt

plt.style.use('default')

# Gridspec_kw creates the space between the two drawings
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 2), gridspec_kw={'wspace': 0.4})

def draw_magnetic_line(ax, start, jump, color, label_text):
    # 1. Main Line with Arrows
    ax.annotate('', xy=(3.5, 0), xytext=(-3.5, 0),
                arrowprops=dict(arrowstyle='<->', color='black', lw=1.5))
    
    # 2. Points and the jump arrow
    end = start + jump
    ax.scatter([start, end], [0, 0], color=['green', 'red'], s=70, zorder=3)
    
    # Jump curve
    connection = "arc3,rad=.5" if jump < 0 else "arc3,rad=-.5"
    ax.annotate('', xy=(end, 0), xytext=(start, 0),
                arrowprops=dict(arrowstyle="->", connectionstyle=connection, 
                color=color, lw=2.5))

    # 3. ATTACH NUMBERS (The Fix)
    # We draw the numbers and ticks manually to ensure zero distance
    ticks = range(-3, 4)
    for x in ticks:
        # Draw the small vertical tick mark
        ax.plot([x, x], [-0.03, 0.03], color='black', lw=1)
        # Place the number EXACTLY under the line
        ax.text(x, -0.08, str(x), ha='center', va='top', fontsize=10)

    # 4. Clean up the standard axis (Hide the default labels)
    ax.set_xlim(-3.8, 3.8)
    ax.set_ylim(-0.4, 0.6)
    ax.axis('off') # Hides all default boxes/ticks

    # 5. Title below the whole structure
    ax.text(0, -0.35, label_text, fontsize=11, fontweight='bold', ha='center', va='top')

# --- Draw Line 1 ---
draw_magnetic_line(ax1, 2, -3, 'dodgerblue', "Addition: 2 + (-3) = -1")

# --- Draw Line 2 ---
draw_magnetic_line(ax2, 1, 2, 'rebeccapurple', "Subtraction: 1 - (-2) = 3")

plt.show()
```

## 3. The Simplified Sign Cheat Sheet
When dealing with multiple signs in a row, use the parity rule:
* **Even number of (-) signs:** Results in **Positive (+)**
* **Odd number of (-) signs:** Results in **Negative (-)**

**Examples:**
* $-(-(-2)) = -2$ (3 signs = Odd)
* $-(-2) = 2$ (2 signs = Even)

### Bridge 1 & 2: Linux Permissions & Cybersecurity (The Revoke/Restore Logic)
**Logic:** In Linux `chmod` or IAM policies, a negative operation is a **Revoke** command. "Subtracting a negative" is like reversing a ban—it restores access.

```python
# Linux/Cyber: Access Control Simulation
permissions = {"read": True, "write": True}

# Operation: Revoke Write (-Write)
permissions["write"] = False 

# Operation: Subtracting a Restriction (Reversing a Ban)
# If 'banned' is a negative state (-1), removing it (- -1) makes it positive
is_banned = True
restore_access = not is_banned # The double-negative flip

print(f"User Permissions: {permissions}")
print(f"Access Restored: {restore_access}")
```

### Bridge 3: Computer Architecture (Integer Arithmetic)
**Logic:** CPUs use **Two's Complement** to handle signed numbers. Subtraction is physically performed by adding the opposite.

```python
# CPU Logic: Subtraction via Addition of Opposites
register_v1 = 10
register_v2 = -5
result = register_v1 + register_v2 # 10 + (-5)

print(f"ALU Result: {result}")
```

### Bridge 4: Software Architecture (State Management)
**Logic:** Managing "Undo" stacks. Every negative move in the state is a "back" command.

```python
# Undo Logic
state_history = [10, 20, 30]
# Subtraction of the last state
current_state = state_history[-1]
undo_move = -10
new_state = current_state + undo_move

print(f"Updated System State: {new_state}")
```

### Bridge 5 & 6: AI & Cloud (Cost & Weight Offsets)
**Logic:** In AI, "Negative Weights" inhibit neurons. In Cloud, "Negative Offsets" represent discounts or resource de-allocation.

```python
# Cloud Billing: Credit Application
monthly_bill = 500
credit_adjustment = -50 # A negative value
final_bill = monthly_bill + credit_adjustment # Adding a negative decreases the total

print(f"Final Cloud Invoice: ${final_bill}")
```
